```python
import pandas as pd

data = pd.read_csv("../input/../input/vinbigdata-chest-xray-abnormalities-detection/train.csv")

meta_data = pd.read_csv("../input/meta-data-diplom/meta_data.csv")
meta_data = meta_data.drop(columns=['Unnamed: 0'])
```

```python
import math
def is_correct_boundaries():
    is_correct = True
    #is not NaN
    i = data.x_min.apply(lambda x: math.isnan(x))
    is_correct = is_correct and all(data.loc[i,'class_id'] == 14)
    i = data.x_max.apply(lambda x: math.isnan(x))
    is_correct = is_correct and all(data.loc[i,'class_id'] == 14)
    i = data.y_min.apply(lambda y: math.isnan(y))
    is_correct = is_correct and all(data.loc[i,'class_id'] == 14)
    i = data.y_max.apply(lambda y: math.isnan(y))
    is_correct = is_correct and all(data.loc[i,'class_id'] == 14)
    
    #all is positive or 0.0
    i = data.x_min.apply(lambda x: math.isnan(x) or x>=0)
    is_correct = is_correct and all(i)
    i = data.x_max.apply(lambda x: math.isnan(x) or x>=0)
    is_correct = is_correct and all(i)
    i = data.y_min.apply(lambda y: math.isnan(y) or y>=0)
    is_correct = is_correct and all(i)
    i = data.y_max.apply(lambda y: math.isnan(y) or y>=0)
    is_correct = is_correct and all(i)
    
    #all records in image sizes
    data_widthout_disease = meta_data[meta_data['class_id'] != 14]
    i = data_widthout_disease.x_min <= data_widthout_disease.width
    is_correct = is_correct and all(i)
    i = data_widthout_disease.x_max <= data_widthout_disease.width
    is_correct = is_correct and all(i)
    i = data_widthout_disease.y_min <= data_widthout_disease.height
    is_correct = is_correct and all(i)
    i = data_widthout_disease.y_max <= data_widthout_disease.height
    is_correct = is_correct and all(i)
    return is_correct

is_correct_boundaries()
```
