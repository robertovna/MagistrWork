```python
from os import listdir
from os.path import isfile, join
import yaml

classes = ['Cardiomegali']
def create_yaml_file(name, classes):
    data = dict(
        train = '/kaggle/input/train-cardiomegali500-class3rad8-illness-cropped/train',
        val   = '/kaggle/input/test-cardiomegali500-class3rad8-illness-cropped/test', 
        names = classes
    )
    with open(name, 'w') as outfile:
        yaml.dump(data, outfile, default_flow_style=False)
        
file_name = 'custom.yaml'
create_yaml_file(file_name, classes)
f = open(file_name, 'r')
print('\nYaml content:')
print(f.read())
```
