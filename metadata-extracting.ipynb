```python
import pydicom
def get_dicom_value_by_tag(dcm, tag_name):
    if tag_name in dcm:
        return dcm[tag_name].value
    else:
        return None
```

```python
sex_key = (0x0010, 0x0040) # пол пациента
age_key = (0x0010, 0x1010) # возраст пациента
weight_key = (0x0010, 0x1030) # вес пациента
size_key = (0x0010,0x1020) # рост пациента
bmi_key = (0x0010,0x1022) # индекс массы тела
allergies_key = (0x0010,0x2110) # аллергия пациента
smoking_key = (0x0010,0x21A0) # статус курения YES NO UNKNOWN
pregnancy_key = (0x0010,0x21C0) # cтатус беременности 01 не беременна 02 возможно 03 да 04 неизвестно

width_key = (0x0028, 0x0011)
height_key = (0x0028, 0x0010)


key_arr =  {'sex': sex_key, 'age': age_key, 'weight': weight_key, 'size': size_key, 'bmi': bmi_key, 
                      'allergies': allergies_key, 'smoking': smoking_key, 'pregnancy': pregnancy_key,
                       'height': height_key, 'width': width_key}
```

```python
import datetime
import pandas as pd

meta_data = pd.read_csv("../input/../input/vinbigdata-chest-xray-abnormalities-detection/train.csv")
train_dir = '../input/vinbigdata-chest-xray-abnormalities-detection/train'
for key in key_arr:
    meta_data[key] = ''
    
start_time = datetime.datetime.now()
for index, row in meta_data.iterrows():
    file_path = train_dir + "/" + row['image_id'] + '.dicom'
    dcm = pydicom.dcmread(file_path, stop_before_pixels=True) # оптимизация - до чтения пикселей останавливаться- не читать весь файл полностью
    for key_name in key_arr:
        meta_data.loc[index, key_name] = get_dicom_value_by_tag(dcm, key_arr[key_name])
        
current_time = datetime.datetime.now()
elapsed_minutes = int((current_time - start_time).total_seconds()) // 60
print('Кол-во минут:', elapsed_minutes)
meta_data.to_csv("meta_data.csv")
```
